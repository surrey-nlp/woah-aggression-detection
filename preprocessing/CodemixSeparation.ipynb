```python
from transformers import pipeline
classify = pipeline(model="l3cube-pune/hing-bert-lid")
```

```python
# Change path of csv
myfile = pd.read_csv("/content/drive/MyDrive/MTechThesis/_10kTweets_masked.csv", dtype=str, on_bad_lines='skip')
print(myfile.head(5))
```

```python
myfile.dtypes
```

```python
tweetscol = list(myfile['TWEET'])
```

```python
x = []

for i in range(0, len(tweetscol)):
  if type(tweetscol[i]) != str:
    x.append(i)

print(x)
```

```python
Langlabs = classify(tweetscol)
```

```python
print(Langlabs[0])
```

```python
new_list = list()
for i in range(0, len(Langlabs)):
  temp_list = list()
  for x in Langlabs[i]:
    temp_list.append(x['entity'])
  new_list.append(temp_list)
```

```python
new_list[:5]
```

```python
print(len(new_list))
```

```python
print(new_list)
```

```python
# 0 --> English, 1 --> Codemixed

final_list = []
for i in range(0, len(new_list)):
  hi = 0
  for x in new_list[i]:
    if x == 'HI':
      hi += 1
  try:
    #print(round(hi/len(new_list[i]), 2))
    if round(hi/len(new_list[i]), 2) >= 0.12:
      final_list.append('1')
    else:
      final_list.append('0')
  except:
    final_list.append('-1')
print(final_list)
```

```python
print(final_list[12])
```

```python
myfile['ENorHI'] = final_list
```

```python
myfile.head(5)
```

```python
en, codemix, m = 0, 0, 0
for i in myfile["ENorHI"]:
  if i == '0':
    en += 1
  elif i == '1':
    codemix += 1
  else:
    m += 1
```

```python
print("EN: ", en)
print("Codemix: ", codemix)
print("Missed: ", m)
```
