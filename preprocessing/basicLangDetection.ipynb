```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Import model



```python
from transformers import pipeline
classify = pipeline(model="papluca/xlm-roberta-base-language-detection", max_length=512, truncation=True)
```

### Read data

### First Half

```python
# Change path of csv
df = pd.read_csv("https://raw.githubusercontent.com/nazianafis/MastersThesis/main/data/FFFallrawdata_HALF1.csv?token=GHSAT0AAAAAAB5DQF2MUBMUS6WBNTI2NIWYY6VI2HQ", dtype = str) #, on_bad_lines='skip')
```

```python
df.head()
```

```python
tweetcol1 = list(df['tweet'])
```

```python
langLabels = classify(tweetcol1)
```

```python
final_labels = [i['label'] for i in langLabels]
```

```python
df['language'] = final_labels
```

```python
print(df.head())
```

```python
for i in range(len(final_labels)):
    if final_labels[i] == 'en' or final_labels[i] == 'hi':
        continue
    else:
        df = df.drop(i)
        
print(df.head(5))
```
